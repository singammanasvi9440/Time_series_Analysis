**Resampling**

Resampling time series data in Python is a common operation when you want to change the frequency of your data, for example, converting daily data to monthly or weekly.

```python
#Montly average stock price
print("Montly Average",df.Close.resample('M').mean())  #monthly
#print("Weakly Average",df.Close.resample('W').mean()) #weekly
#df.Close is selecting Close column from data frame and resample montly wise by finding mean
```

```python
df.Close.mean()
```

```python
%matplotlib inline
df.Close.resample('M').mean().plot()
```

```python
df.Close.resample('Q').mean().plot(kind="bar") #Q is quaterly
```

```python
#generate dates
#here freq='B which means business days,if you give D its daily
rng=pd.date_range(start="6/1/2017",end='6/30/2017',freq='B')
rng
```
